```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/owid/covid-19-data/refs/heads/master/public/data/owid-covid-data.csv'
df = pd.read_csv(url)
df
```

```python
df.shape #dimensioni dataset
```

```python
df.info() #panoramica coincisa DF
```

```python
df1 = df.dropna(subset = ['continent', 'location']) #ogliere dati nulli da 'continent' e 'location':  
#le righe dove la colonna 'continent' è nulla hanno dei valori aggregati, in quanto in 'location' hanno 'Wolrd', 'Europe' e 'High Income'
df1
```

```python
df_max = df.groupby(['continent','location'])['total_cases'].agg(max) #trovare massimo per località raggruppando per continente: 
df_max
```

```python
df_max_per_continent = df_max.groupby(level = 'continent').sum() 
#sommare le singole località per trovare il max 'total cases' per continente, mediante il multiindex:

df_max_per_continent
```

```python
tot_mondiale = df_max_per_continent.sum() #sommiamo i tot cases per continente per avere il totale nel mondo:
tot_mondiale
```

```python
percentuale = (df_max_per_continent/tot_mondiale)*100 #troviamo la percentuale rispetto al totale mondiale del numero di casi:

percentuale
```

```python
# per maggiore comprensione uniamo in un unico dataframe:
# creo un nuovo DF
df_finale = pd.DataFrame(df_max_per_continent)
df_finale

#aggiungo la colonna della percentuale rispetto al mondo:
df_finale['percentuale_mondiale'] = (df_finale['total_cases']/tot_mondiale)*100

#aggiungo colonna di riepilogo per il tot mondiale sia numerico che percentuale
df_finale.loc['Mondo Complessivo'] = [tot_mondiale, 100.0]

#sistemo la percentuale affinché abbia solo due numeri dopo la virgola:
df_finale['percentuale_mondiale'] = df_finale['percentuale_mondiale'].round(2)

#stampo risultato finale
df_finale
```

```python
#convertire dati colonna date da stringa a dati temporali:
df['date'] = pd.to_datetime(df['date'])
df
```

```python
# filtriamo il Df per 'location' == 'Italy' e anno == 2022
df_italy_2022 = df[(df['location'] == 'Italy') & (df['date'].dt.year == 2022)]
df_italy_2022
```

```python
# togliamo ora i valori nulli e con misurazione 0 nella colonna new_cases:
df_italy_2022 = df_italy_2022.dropna(subset=['new_cases'])
df_italy_2022 = df_italy_2022[df_italy_2022['new_cases']>0]
df_italy_2022
```

```python
# creiamo ora grafico per mostrare l'evoluzione dei casi totali dall'inizio alla fine dell'anno:

import seaborn as sns #per creare grafico
import matplotlib.pyplot as plt #per sistemare grafico - letta apposita documentazione
import matplotlib.ticker as ticker #per togliere la notazione scientifica - visto con Gemini e poi visto sulla documentazione

grafico_total_cases = sns.lineplot(data=df_italy_2022, x='date', y='total_cases')
grafico_total_cases.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}')) #per togliere notazione scientifica e mettere separatore migliaia

# aggiungo ora i titolo con matplotlib.pylot
plt.title('Andamento Contagiati Totali Italia 2022')
plt.xlabel('Mese')
plt.ylabel('Numero Totale di Casì')

plt.show()

```

```python
# creiamo ora grafico per il numero di nuovi casi rispetto alla data:

#lascio lo stesso codice cambiando soltanto da total_cases a new_cases e sistemando i titoli
grafico_new_cases = sns.lineplot(data=df_italy_2022, x='date', y='new_cases')
grafico_new_cases.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}')) #togliere notazione scientifica e mettere separatore migliaia

# aggiungo ora i titolo con matplotlib.pylot
plt.title('Andamento Nuovi Contagiati Italia 2022')
plt.xlabel('Mese')
plt.ylabel('Numero di Casì')

plt.show()
```

```python
# filtriamo il dataframe per Germany, Italy e France nel periodo maggio 2022 e aprile 2023
nazioni = ['Italy', 'France', 'Germany']
periodo_iniziale = '2022-05-01'
periodo_finale = '2023-04-30'

filtro_nazioni = df.loc[:, 'location'].isin(nazioni)
filtro_data = (df.loc[:, 'date'] >= periodo_iniziale) & (df.loc[:, 'date'] <= periodo_finale)

# ifg sta per Italy, France e Germany
df_ifg = df.loc[filtro_nazioni &filtro_data]
df_ifg
```

```python
# creiamo ora il boxplot come da consegna:
grafico_icu = sns.boxplot(data = df_ifg, x='location', y='icu_patients')
grafico_icu
```

```python
# commento:
# Dall'analisi del boxplot relativo al periodo maggio 2022 - aprile 2023 notiamo che:
# -la Germania ha valori assoluti più elevati in termini di pazienti in terapia intensiva appunto, dovuto sicuramente alla maggiore popolazione rispetto Francia (che è subito dopo) e Italia  (la quale mostra volumi nettamente inferiori). Questo dato, oltre che dalla popolazione sarà stato influenzato anche dalle diverse capacità ospedaliere dei diversi paesi;
# -la Germania prima e la Francia subito dopo mostrano delle scatole più estese rispetto all'Italia. Questo, in linea con il maggior numero di valori assoluti, ci fa capire che questi 2 stati hanno ricevuto un carico maggiore nei reparti di terapia intensiva durante questo arco temporale;
# -l'Italia mostra una scatola più compressa e una mediana molto più bassa rispetto Germania e Francia. Questo ci fa capire che in questo intervallo temporale la pressione in terapia intensiva in Italia è rimasta molto più costante e con numeri generali molto più bassi rispetto le altre due. Questo può essere dovuto, o dal fatto che l'Italia abbia una popolazione minore rispetto le altre due, o al fatto che non ci sono stati particolari picchi durante questo arco temporale (lo si nota anche dagli outlier molto più contenuti in ambo i sensi), o ad entrambi i due fattori;
# -per quanto riguarda gli outlier, la Germania presenta dei baffi verso l'alto molto più grandi rispetto Italia e Francia. Questo sarà dovuto sicuramente principalmente da una maggiore popolazione e/o da particolari picchi che hanno portato sotto maggiore stress i reparti di terapia intensiva.
```

```python
# filtriamo per le nazioni Italia, Spagna, Francia e Germania per tutto il 2021:
nazioni = ['Italy','Spain', 'France', 'Germany']
inzio_periodo = '2021-01-31'
fine_periodo = '2021-12-31'

filtro_nazioni = df.loc[:,'location'].isin(nazioni)
filtro_data = (df.loc[:, 'date'] >= inzio_periodo) & (df.loc[:, 'date'] <= fine_periodo)

df_nuovo = df.loc[filtro_nazioni&filtro_data]
df_nuovo
```

```python
# sommare il numero di ospitalizzati raggruppando appunto per nazione - Maniera Numerica:
df_hosp = df_nuovo.groupby('location')['hosp_patients'].sum()
df_hosp
```

```python
#Grafico a barre: 
grafico_finale = df_hosp.plot(kind='bar', color=['skyblue', 'orange', 'green', 'red'])
grafico_finale.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}')) #generato sempre con Gemini per togliere notazione scientifica dall'asse y
```
